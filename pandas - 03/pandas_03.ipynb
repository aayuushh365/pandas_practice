```python
# use converter to change data in a column
import pandas as pd

def convert_people_cell(cell):
 if cell =="n.a.":
  return 'sam walton'
 return cell

def convert_eps_cell(cell):
 if cell == "not available":
  return None
 return cell

def convert_revenue_cell(cell):
 if cell == '87':
  return '69'
 return cell

sc = pd.read_csv("stock_data.csv", converters= {
 'people': convert_people_cell ,
 'eps': convert_eps_cell ,
 'revenue': convert_revenue_cell,
 })
sc
```

```python
# writing code back to excel file
sc.to_excel("new.xlsx", sheet_name = "stocks", startrow=1 , startcol=2)
```

```python
# if you have 2 excel files and you want to use them in the same spreadsheet. 
df_stocks = pd.DataFrame({
    'tickers':['GOOGL','WMT','MSFT'],
    'price':[845,65,64],
    'pe':[30.37,14.26,30.97],
    'eps':[27.82,4.61,2.12]
})
df_weather = pd.DataFrame({
    'day':['1/1/2017','1/2/2017','1/3/2017'],
    'temperature':[32,35,28],
    'event':['Rain','Sunny','Snow'] })
```

```python
with pd.ExcelWriter('stocks_weather.xlsx')as writer:
   df_stocks.to_excel(writer,sheet_name="stocks")
   df_weather.to_excel(writer,sheet_name="weather")
```
